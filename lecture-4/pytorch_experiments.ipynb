```python
import torch
```

```python
transitions = [1,2,3,4,5]
x = torch.empty(size=(len(transitions), 1))

for i,val in enumerate(transitions):
    x[i] = val
```

```python
x = torch.empty(size=(1, 1))
```

```python
for i,val in enumerate(transitions):
    if x.shape[0] == 1:
        x[0] = val
    else:
        x = torch.cat(x, val)
```
